```python
import pandas as pd
from sklearn.model_selection import train_test_split
import datetime
from sklearn.metrics import mean_squared_error
import joblib
```

```python
data_path = "F:\CrimePrediction\data\chicago_data.csv"

data_file = pd.read_csv(data_path)
data_file = data_file.dropna(axis=0)  # drop null values on the row
```

```python
data_file["Date"] = pd.to_datetime(data_file["Date"])
data_file["Year"] = data_file["Date"].dt.year 
data_file["Month"] = data_file["Date"].dt.month 
data_file["DayOfTheWeek"] = data_file["Date"].dt.dayofweek
data_file["Hour"] = data_file["Date"].dt.hour
data_file
```

```python
features = ["DayOfTheWeek", "Month", "Hour", "Year"]

target = ["Latitude", "Longitude"]

X = data_file[features]
Y = data_file[target]

x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=42, test_size=0.3)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"estimation_Num": [1, 5, 25, 50, 100], "esimation_maxdepth": [1, 2, 5, 10, 25, 50]}
grid_search = GridSearchCV(estimator=RandomForestRegressor(),param_grid=param_grid, cv=5)
grid_search.fit(x_train, y_train)
```

```python
best_param = grid_search.best_params_
best_model = RandomForestRegressor(n_estimators=best_param["estimation_Num"], max_depth=best_param["esimation_maxdepth"])
best_model.fit(x_train, y_train)
```
